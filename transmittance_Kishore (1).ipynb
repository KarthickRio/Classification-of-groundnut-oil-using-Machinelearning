#### Calculating the mean of the water data to find the transmittance by slicing

```python
# slicing the water data by index_loc
Water_data=df.iloc[30:33]
Water_data.reset_index()
```

```python
#mean is calculated by numpy and axis is defined for row

import numpy as np
Water_average=Water_data.mean(axis=0)
Water_averaged=(Water_average.reset_index().T)
```

```python
Water_averaged
```

#### List iteration

```python
#The mean is averaged and converted to list for iteration
x=list(Water_averaged.iloc[1])
x
```

## For loop initilizing,iteration and appending to a dataframe and transposing it

```python
num=pd.DataFrame()
list_of_dfs = [] # Create an empty list to store dataframes
for i in range(0,129):
    j=i-0
    s=df.iloc[:,(i-1):i]
    k=x[j-1]
    a=s/k
    list_of_dfs.append(pd.DataFrame(a).T) # Append transposed dataframe to the list
    # print(num) # Commenting out print statement for cleaner output

num = pd.concat(list_of_dfs, ignore_index=True) # Concatenate all dataframes in the list
```

```python
#The value is transposed to return a normal dataframe
f_df=num.T
```
